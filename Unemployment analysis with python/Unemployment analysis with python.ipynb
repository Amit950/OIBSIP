```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Unemployment in India.csv")

# Remove leading and trailing whitespaces from column names and values
df.columns = df.columns.str.strip()
df['Date'] = df['Date'].str.strip()

# Convert the 'Date' column to datetime with the correct format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Calculate the average unemployment rate for each year
df['Year'] = df['Date'].dt.year
df['Avg_Unemployment_Rate'] = df.groupby('Year')['Estimated Unemployment Rate (%)'].transform('mean')

# Plot the average unemployment rate over time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Avg_Unemployment_Rate'], marker='o')
plt.title('Average Unemployment Rate in India')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
